Escreva uma função chamada **compara** que receba como parâmetro duas listas duplamente ligadas. Ela deve retornar True caso as duas listas duplamente ligas passadas como parâmetro sejam iguais e False caso
contrário. Duas listas são consideradas iguais tanto quando os elementos estão na mesma ordem ou em ordem contrária. Exemplo: A = [0,1,2,3,4], B = [0,1,2,3,4] e C = [4,3,2,1,0] são listas iguais.
Você pode usar o código a lista duplamente ligada criado na disciplina. 

```python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class Node:
    """
    Noh de uma lista duplamente ligada
    """
    def __init__(self, data = None, next = None, prev = None):
        self.data = data
        self.next = next # apontador para o noh da frente
        self.prev = prev # apontador para o noh de tras

class Lista:

    def __init__(self, data  = None):
        self.head = None
        self.tail = None
        self.size = 0

    def insertHead(self, data):
        """
        Adiciona um elemento no inicio da lista
        """
        
        # cria um novo noh com o dado informado
        tempNode = Node(data)

        # se a lista esta vazia, o element head
        # ira receber o novo noh.
        # Nesse caso,o elemento tail devera apontar para 
        # o mesmo noh do elemento head        
        if self.size==0:
            self.head = tempNode
            self.tail = tempNode
            
        else:
            
            # o proximo elemento na lista a frente do novo noh sera o noh head antigo
            tempNode.next = self.head
            
            # o elemento atras do noh head antigo sera o novo noh
            self.head.prev = tempNode
            
            # o novo noh head sera o novo noh
            self.head = tempNode
            
        # incrementa a quantidade de elementos da lista
        self.size += 1
        
    def insertTail(self, data):
        """
        Adiciona um elemento no final da lista
        """

        # cria um novo noh com o dado informado
        tempNode = Node(data)
            
        # se a lista esta vazia, o element tail
        # ira receber o novo noh.
        # Nesse caso,o elemento head devera apontar para 
        # o mesmo noh do elemento tail
        if self.size==0:
            self.tail = tempNode
            self.head = self.tail
            
        else:
            
            # o tail antigo ficara atras do novo noh e, por isso,
            # o apontador prev do novo noh deve apontar para o tail antigo
            tempNode.prev = self.tail
            
            # o noh tail sera substituido pelo novo noh e, por isso,
            # seu apontador next deve apontar para o novo noh
            self.tail.next = tempNode
            
            # o novo noh tail sera o novo noh
            self.tail = tempNode
        
        self.size += 1

    def __str__(self):
        """
        Retorna uma string com as informacoes da lista
        quando o objeto e chamado dentro de um print() 
        ou dentro de um str().
        
        A lista nao possui restricao de acesso aos seus elementos. 
        Por isso, ela nao precisa ser destruida para que seus elementos sejam 
        visitados como na fila e na pilha
        """

        strfila = '[ '
        
        if self.size > 0:
            
            # comecao pelo head
            tempNode  = self.head
            valor = tempNode.data
        
            # guarda o valor do head em uma string
            strfila += ' ' + str(valor)
            
            # percorre o resto da lista ate que o apontador next do noh seja nulo
            # o que indica que chegou no tail
            while tempNode.next is not None:
                
                # vai para o proximo noh
                tempNode = tempNode.next
                
                valor = tempNode.data 
                
                # guarda o valor em uma string
                strfila += ' ' + str(valor)
            
        strfila += ' ]'
        
        return strfila
      
def compara(lista1,lista2):
    
    verifica = False

    #Se o tamanho das listas forem diferentes, as listas ja não são iguais
    if lista1.size != lista2.size:
        return False

    #Se o primeiro item da lista for igual ao primeiro item da segunda lista
    if lista1.head.data == lista2.head.data:
        
        #Inicializando os ponteiros no inicio
        ponteiro1 = lista1.head
        ponteiro2 = lista2.head

        #Loop enquanto o ponteiro da lista1 for diferente de vazio
        while ponteiro1.next is not None:
            
            #Compara item por item da lista 
            if(ponteiro1.data == ponteiro2.data):
                verifica = True
            else:
                return False
            
            #Passando para a proxima posicao dos dois ponteiros
            ponteiro1 = ponteiro1.next
            ponteiro2 = ponteiro2.next
            
    #Se o primeiro valor da lista 1 for igual ao ultimo valor da lista dois, ou vice-versa
    if lista1.head.data == lista2.tail.data:
        
        #Incializando um ponteiro no inicio e outro no final
        ponteiro1 = lista1.head
        ponteiro2 = lista2.tail

        #Loop enquanto o ponteiro da lista1 for diferente de vazio
        while ponteiro1.next is not None:
            
            #Compara item por item da lista 
            if(ponteiro1.data == ponteiro2.data):
                verifica = True
            else:
                return False
                
            
            #Passando para a proxima posicao dos dois ponteiros
            ponteiro1 = ponteiro1.next
            ponteiro2 = ponteiro2.prev
    
    return verifica


##########################################################################

# testando a função criada
lista1 = Lista()
lista1.insertHead(5)
lista1.insertHead(9)
lista1.insertTail(3)
lista1.insertTail(4)

lista2 = Lista()
lista2.insertHead(3)
lista2.insertHead(4)
lista2.insertTail(5)
lista2.insertTail(9)

lista3 = Lista()
lista3.insertHead(5)
lista3.insertHead(9)
lista3.insertTail(1)
lista3.insertTail(4)

compLista1_lista1 = compara(lista1, lista1)
compLista1_lista2= compara(lista1, lista2)
compLista1_lista3= compara(lista1, lista3)


print('Resultado retornado:')
print('\tLista 1 == Lista 1?: ', compLista1_lista1)
print('\tLista 1 == Lista 2?: ', compLista1_lista2)
print('\tLista 1 == Lista 3?: ', compLista1_lista3)

print('\n' + 20*'-')
print('Resultado esperado:')
print('\tLista 1 == Lista 1?: True')
print('\tLista 1 == Lista 2?: True')
print('\tLista 1 == Lista 3?: False')
```

## Exercício 2

Crie uma função recursiva que receba uma lista duplamente ligada e retorne como resultado uma nova lista duplamente ligada com os valores em ordem invertida. Exemplo: se a lista passada como entrada for [1,2,3,4,5], ele deve retornar [5,4,3,2,1]. Você pode usar o código a lista duplamente ligada criado na disciplina.

```python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
#No de uma lista duplimente ligada
class Node:

    #Função iniciadora
    def __init__(self, data=None, next=None, prev=None):
        
        #Dados enviados pelo usuario
        self.data = data

        #Ponteiro para a frente do no
        self.next = next

        #Ponteiro para o noh de tras 
        self.prev = prev

#Classe que manipula a lista duplamente ligada
class Lista:

    #Função iniciadora
    def __init__(self, data=None):
        
        #Posicao inicial da lista
        self.head = None
        
        #Posição final da lista
        self.tail = None

        #Tamanho da lista
        self.size = 0

    #Função para adicionar o item no começo da lista
    def insertHead(self, data):

        #Cria um no auxiliar com os dados informados
        tempNode = Node(data)

        #Se a lista estiver vazia (primeira inserção)
        if self.size == 0:

            #O primeiro indice da lista (header) recebe o no auxiliar 
            self.head = tempNode

            #E o ponteiro tail (aponta para a parte de tras da lista) aponta para o elmento auxiliar
            self.tail = tempNode

        #Se a lista ja tiver conteudo
        else:

            #O proximo elemento na lista a frente do novo noh head antigo
            tempNode.next = self.head

            #O elemento atras do noh head antigo sera o novo noh
            self.head.prev = tempNode

            # o novo noh head sera o novo noh
            self.head = tempNode

        #Incrementa a quantidade de elementos da lista
        self.size += 1

    #Função para inserir dados no final da lista 
    def insertTail(self,data):

        #Cria um novo noh com o dado informado pelo usuario
        tempoNode = Node(data)

        #Se a lista esta vazia (primeira inserção)
        if self.size == 0:
            
            #O elemento tail recebe o novo noh
            self.tail = tempoNode

            #O elemento head deve apontar para o mesmo noh que aponta o tail
            self.head = self.tail

        else:

            #O ponteiro prev do novo noh aponta par o tail antigo
            tempoNode.prev = self.tail

            #o novo noh sera substituido pelo novo noh, por isso
            #O apontador next deve apontar para o nobo noh
            self.tail.next = tempoNode

            #O novo noh tail sera o novo noh
            self.tail = tempoNode
        
        #Incrementando o tamanho
        self.size += 1

    #Função responsavel por retornar uma string com as informações da lista 
    def __str__(self):

            strfila = '['
            
            if self.size > 0:
                
                # comecao pelo head
                tempNode  = self.head
                valor = tempNode.data
            
                # guarda o valor do head em uma string
                strfila += '' + str(valor)
                
                # percorre o resto da lista ate que o apontador next do noh seja nulo
                # o que indica que chegou no tail
                while tempNode.next is not None:
                    
                    # vai para o proximo noh
                    tempNode = tempNode.next
                    
                    valor = tempNode.data 
                    
                    # guarda o valor em uma string
                    strfila += ',' + str(valor)
                
            strfila += ']'
            
            return strfila

#Função responsavel, por retornar a lista invertida
def inverte(lista,tempNode=None,str_lista = '['):
    
    #Se for a primeira vez que o loop ira rodar 
    if tempNode == None:

        #Inicializa o ponteiro auxiliar para o ultimo elemento da lista
        tempNode = lista.tail

        #E a variavel auxiliar com o valor que este ponteiro esta apontando
        auxValor = tempNode.data

        #Iniciando a string responsavel por armazenar os dados da lista
        str_lista += '' + str(auxValor)

    else:
        #Se não for o primeiro loop, o valor recebe o valor que o ponteiro esta apontando
        auxValor = tempNode.data
        
        #Guardando o valor na strig que sera retornada depois
        str_lista += ',' + str(auxValor)

    #Se o valor auxiliar é igual ao primeiro elemento da lista, toda a lista foi percorrida
    if auxValor == lista.head.data:

        #Fechando a lista com o parenteses
        str_lista += ']'

        #Por isso paramos o processo
        return str_lista

    else:

        #Se não, o ponteiro aponta para o elemento anterior da lista
        tempNode = tempNode.prev

        #auxValor recebe o valor que o ponteiro esta apontando
        auxValor = tempNode.data

        #Chamamos a função novamente alterando a posição do ponteiro, e o valor que esta apontando
        return inverte(lista,tempNode,str_lista)   

##########################################################################

# testando a função criada
lista1 = Lista()
lista1.insertHead(5)
lista1.insertHead(9)
lista1.insertTail(3)
lista1.insertTail(4)

lista2 = inverte(lista1)

print('Resultado retornado:')
print('\tLista invertida: ', lista2)

print('\n' + 20*'-')
print('Resultado esperado:')
print('\tLista invertida: [4,3,5,9]')
```

```python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class Node:
    
    def __init__(self, data):
        
        self.data = data
        self.right = None
        self.left = None

        #Para calcular o balanceamento da arvore precisa adicionar a variavel altura
        #que sera incrementada toda vez que houver uma nova inserção
        self.height = 1
        
class Tree:
    """
    Arvore binaria de busca - implementacao dinamica
    """

    def __init__(self):
        self.root = None
        
    def insert(self, data, node = None):
        """
        Insere um novo noh. Considera a regra de arvores
        binarias de busca em que a subarvore esquerda de um determinado noh 
        deve ter uma chaves menores que a desse noh, enquanto que a subarvore direita 
        deve ter chaves maiores.
        Alem disso, como essa arvore eh AVL, a cada insercao, 
        caso a arvore fique desbalanceada, deve ser feito o balanceamento.
        """

        # se a arvore ainda nao possui noh raiz
        if self.root == None:
            self.root = Node(data)
            node = self.root
            return node
        else:
            
            if node is None: 
                node = self.root
        
            # se o valor do novo noh for menor que o valor do noh atual, insere na esquerda
            if data < node.data:
                
                if node.left is None:
                    
                    # insere um novo noh com o valor desejado na esquerda
                    node.left = Node(data)
                    
                else:
                    
                    # se ja existe um noh na esquerda, nao podemos inserir o novo noh
                    # diretamente. Precisamos chamar a funcao insert recursivamente no 
                    # noh da esquerda
                    node.left = self.insert( data, node.left )
                
                
            # se o valor do novo noh for maior ou igual que o valor do noh atual, insere na direita
            else:
                
                if node.right is None:
                    
                    # insere um novo noh com o valor desejado na direita
                    node.right = Node(data)
                    
                else:

                    # se ja existe um noh na direita, nao podemos inserir o novo noh
                    # diretamente. Precisamos chamar a funcao insert recursivamente no 
                    # noh da direita
                    node.right = self.insert( data, node.right )  
                    
                    
            #A partir de uma nova inserção, realizamos o incremento da altura da arvoreFaltura
            
            #Se existir um nó na esquerda da raiz 
            if node.left is not None:
                #Pega a altura do nó da esquerda
                left_height = node.left
                left_height = left_height.height
            else:
                #Se não, altura da esquerda é 0
                left_height = 0
            
            #Se existir um nó na direita da raiz
            if node.right is not None:
                right_height = node.right
                right_height = right_height.height
            else:
                right_height=0

            #Depois de pegar a altura das subarvores, somamos o valor da raiz (1) com a altuda da maior subárvore
            maior = 0
            
            #Verificanndo qual subarvore e mais alta
            if(left_height > 0):
                maior = left_height
            else:
                maior = right_height
            
            #Atribuindo o valor da arvore
            node.height = 1 + maior

            self.root = node
            return node
                         
    def strInorder(self, node = -1, info = ''):
        """
        Retorna uma string com os valores da arvore obtidos apos 
        o percurso "Em Ordem"
        """
        
        if self.root is None:
            return ' '
            
        else:
            
            if node==-1:
                node = tree.root
            
            if node.data is not None:
                
                if node.left is not None: 
                    info += self.strInorder(node.left)
                    
                info += ' ' + str(node.data) #print(data)
                
                if node.right is not None:
                    info += self.strInorder(node.right)
                    
                return info
            else:
                return info
                   
    def strPreorder(self, node = -1, info = ''):
        """
        Retorna uma string com os valores da arvore obtidos apos 
        o percurso "Pre Ordem"
        """
        
        if self.root is None:
            return ' '
            
        else:
            
            if node==-1:
                node = tree.root
            
            if node.data is not None:
                
                info += ' ' + str(node.data)
                
                if node.left is not None: 
                    info += self.strPreorder(node.left)
                
                if node.right is not None:
                    info += self.strPreorder(node.right)
                    
                return info
            else:
                return info
            
    def strPostorder(self, node = -1, info = ''):
        """
        Retorna uma string com os valores da arvore obtidos apos 
        o percurso "Pre Ordem"
        """
        
        if self.root is None:
            return ' '
            
        else:
            
            if node==-1:
                node = tree.root
            
            if node.data is not None:
                              
                if node.left is not None: 
                    info += self.strPostorder(node.left)
                
                if node.right is not None:
                    info += self.strPostorder(node.right)
                    
                info += ' ' + str(node.data)
                    
                return info
            else:
                return info
    
    def buscar(self, searchedData, node = -1):
    
        if node == -1:
            node = tree.root
            
        if node.data is not None:
            
            if searchedData == node.data:
                return True
            
            elif searchedData < node.data:
                
                if node.left is not None: 
                    return self.buscar( searchedData, node.left )
                else: 
                    return False
            else:
                if node.right is not None:
                    return self.buscar( searchedData, node.right )
                else:
                    False
        else:
            return False 

    def verificaBalanceamento(self, node = -1, estaBalanceada = True, balanceamento = ''):

        #Verifica se existe uma raiz
        if self.root is None:

            #Se não existe o processo ja para aqui
            return estaBalanceada, balanceamento
        
        #Se o nó for igual ao valor inicial ou seja não existir nó
        if node == -1:

            #Aponta o nó para o valor da raiz
            node = self.root
        
        #Se o nó não tem valor
        if node.data is None:
            
            #O fator de balanceamento não existe tambem
            balance = 0
            balanceamento  += " " + str(balance) + "( "
        
        #Se exisitir nó 
        else:
            #Pega a altura dos dois lados da arvore
            #Se existir um nó na esquerda da raiz 
            if node.left is not None:
                #Pega a altura do nó da esquerda
                left_height = node.left
                left_height = left_height.height
            else:
                #Se não, altura da esquerda é 0
                left_height = 0
            
            #Se existir um nó na direita da raiz
            if node.right is not None:
                right_height = node.right
                right_height = right_height.height
            else:
                right_height = 0
            
            #Calcula o fator de balanceamento que seria a subtração da altura da esquerda com a altura da direita
            balance = left_height - right_height

            #Se o fator de balancemanto estiver entre 1, 0 e -1 a arvore esta balanceada
            if balance == 1 or balance == -1 or balance == 0:
                estaBalanceada = True
            else:
                estaBalanceada = False
            
            #Concatena na string que sera retornada o valor de balanceamento
            balanceamento += " " + str(balance) + "( "

            #Se existir valor na esquerda
            if node.left is not None:
                
                #Chama a função novamente para descobrir se a subarvore da esquerda esta balanceada e o seu fator de balanceamento
                left_balance, str_left_balance = self.verificaBalanceamento(node.left)
                balanceamento += str_left_balance

                #Se a sub-arvore da esquerda estiver desbalanceada
                if left_balance == False:

                    #Atribui a varivel False
                    estaBalanceada = False
            
            #Se existir valor na direita
            if node.right is not None:

                #Chama a função novamente para descobrir se a subarvore da direita esta balanceada e o seu fator de balanceamento
                right_balance, str_right_balance = self.verificaBalanceamento(node.right)
                balanceamento += str_right_balance

                #Se a sub-arvore da direita estiver desbalanceada
                if right_balance == False:

                    #Atribui a varivel False
                    estaBalanceada = False
            
            balanceamento += " )" 


        #Retorna se a arvore esta balanceada e a ordem do balanceamento
        return estaBalanceada, balanceamento

#---------------------    
# testando a arvore

tree = Tree()
tree.insert(17)
tree.insert(6)
tree.insert(35)
tree.insert(4)
tree.insert(5)
tree.insert(48)

estaBalanceada, balanceamento = tree.verificaBalanceamento()

print('Resultado retornado:' )
print('\tEstá balanceada: ', estaBalanceada)
print('\tBalanceamento: %s' %(balanceamento))

print('\n' + 20*'-')
print('\nResultado esperado:')
print('\tEstá balanceada: False')
print('\tBalanceamento: 1( 2( -1( 0( ) ) ) -1( 0( ) ) )')
```
